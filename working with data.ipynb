## Task 5

```python
# Number of objects in test sample
n = 500

# Number of objects with prediction higher than expected
n_high = 400

# Number of objects with prediction lower than expected
n_low = n - n_high

# Difference in prediction for objects with prediction higher than expected
diff_high = 0.5

# Difference in prediction for objects with prediction lower than expected
diff_low = 0.7

# Omer's MSE
mse_omer = (n_high * diff_high**2 + n_low * diff_low**2) / n

# Limor's MSE
# Calculating the constant value that Limor uses to change all the answers
constant = (n_high * diff_high - n_low * diff_low) / n
# Limor's MSE is 0 as she found the smallest error
mse_limor = (n_high * (diff_high - constant)**2 + n_low * (diff_low - constant)**2) / n

print("Omer's MSE: ", mse_omer)
print("Limor's MSE: ", mse_limor)
```
